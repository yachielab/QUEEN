## Golden Gate Assembly

This set contains a destination vector pGGAselect (chlor) and 24 fragments encoding part of the lac operon in pUC57 (amp) holding plasmids. The destination vector pGGAselect can be used as a cloning vector for Golden Gate Assembly using BsaI, BsmBI, and BbsI. The 4 bp overhang ends of 24 fragments of lac cassettes assemble with high fidelity and efficiency, with a design based on the comprehensive profiling of the ligation fidelity of T4 DNA ligase. This set can be used as a positive control for a complex assembly using BsaI via blue/white screening as E.coli cells that carry the correctly assembled lac cassettes form blue colonies on an LB plate with X-gal.

*reference: https://www.addgene.org/kits/cantor-golden-gate-assembly/#kit-contents*  
addgene_ids used in the following example: 195714, 195715, 195716, 195717, 195718, 195719, 195720, 195721, 195722, 195723, 195724, 195725, 195726, 195727, 195728, 195729, 195730, 195731, 195732, 195733, 195734, 195735, 195736, 195737, 195738

```python
destination   = QUEEN(record="195714", dbtype="addgene")
entry_ids     = [195715, 195716, 195717, 195718, 195719, 195720, 195721, 195722, 195723, 195724, 195725, 195726, 195727, 195728, 195729, 195730, 195731, 195732, 195733, 195734, 195735, 195736, 195737, 195738]
#random.shuffle(entry_ids)
entry_vectors = [] 
for aid in entry_ids:
    aentry = QUEEN(record="{}".format(aid), dbtype="addgene")
    sites  = aentry.searchsequence(cs.lib["BsaI"].seq, quinable=0)
    aentry.setfeature({"start":sites[0].end+1, "end":sites[-1].start-1, "qualifier:label":"ins-{}".format(int(aid)-195714)})
    entry_vectors.append(aentry)
```

```python
construct = goldengate_assembly(destination, entry_vectors, "BsaI", product="construct")
construct.outputgbk("golden_gate_construct.gbk")
```
