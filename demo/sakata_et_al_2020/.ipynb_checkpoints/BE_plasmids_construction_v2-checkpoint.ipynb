```python
description1 = "The N-terminus of Target-AID was amplified from pcDNA3.1_pCMV-nCas-PmCDA1-ugi pH1-gRNA(HPRT) \
(Addgene 79620) using primer pairs RS045/HM129."
RS045 = "GAGAGCCGCCACCATGGCACCGAAGAAGAAGCG"
HM129 = "CTGGGGCACGATATGATCCACGTCGTAGTCGGAGA"
FW    = pCMV_nCas_PmCDA1_ugi.finddna(RS045[-18:], key_attribute="sequence") #Search FW primer binding region
RV    = pCMV_nCas_PmCDA1_ugi.finddna(HM129[-18:], key_attribute="sequence") #Search RV primer binding region
frag1 = cropdna(pCMV_nCas_PmCDA1_ugi, *FW, *RV, process_description=description1)
editdna(pCMV_nCas_PmCDA1_ugi, key_attribute="sequence", query=RS045[-18:]+frag1.seq+flipdna(HM129[-18:]), target_attribute="feature_id", operation=createattribute("f1"), new_copy=False)
editdna(pCMV_nCas_PmCDA1_ugi, key_attribute="feature_id", query="f1", target_attribute="qualifier:label", operation=createattribute(" "), new_copy=False)
editdna(pCMV_nCas_PmCDA1_ugi, key_attribute="feature_id", query="f1", target_attribute="strand", operation=replaceattribute(0), new_copy=False)
frag1 = modifyends(frag1, RS045, flipdna(HM129), process_description=description1) #Simulate PCR
#Add annotation
editdna(frag1, key_attribute="sequence", query=None, target_attribute="feature_id", operation=createattribute("f1"), new_copy=False)
editdna(frag1, key_attribute="feature_id", query="f1", target_attribute="qualifier:label", operation=createattribute("fragment-1"), new_copy=False)
editdna(frag1, key_attribute="feature_id", query="f1", target_attribute="strand", operation=replaceattribute(0), new_copy=False)
editdna(frag1, key_attribute="sequence", query=RS045, target_attribute="feature_id", operation=createattribute("p1"), new_copy=False)
editdna(frag1, key_attribute="feature_id", query="p1", target_attribute="qualifier:label", operation=createattribute("RS045"), new_copy=False)
editdna(frag1, key_attribute="sequence", query=HM129, target_attribute="feature_id", operation=createattribute("p2"), new_copy=False)
editdna(frag1, key_attribute="feature_id", query="p2", target_attribute="qualifier:label", operation=createattribute("HM129"), new_copy=False)
```
